```python
# 성공 소스. 자바 api 사이트로 들어가서 팝업창을 제거하고, 검색하여 목록을 모아 선택하게 한 후, 선택한 페이지로 이동합니다.

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time, lxml
from selenium.webdriver.common.by import By

root = 'https://docs.oracle.com/en/java/javase/17/docs/api/'
# 팝업 전담 일진 : 거지같은 팝업쉐킷 제거 성공!!!!!! 안녕 지겨웠고 다신 보지 말자!
def kill_iframe(driver):
    iframe = driver.find_element_by_class_name('truste_popframe')
    # 이게 키포인트! iframe에 포커스하면 iframe 안의 엘리먼트를 찾을 수 있습니다!
    driver.switch_to.frame(iframe)
    target = driver.find_element_by_link_text('모두 거절')
    target.click()
    time.sleep(5)
    target = driver.find_element_by_link_text('닫기')
    target.click()
    
def clear_search(driver):
    reset_btn = driver.find_element_by_id('reset-button')
    reset_btn.click()
    
def text_on_search_bar(driver, query):
    search = driver.find_element_by_id('search-input')
    search.click()
    search.send_keys(query)
    time.sleep(3)

chrome_path = './resource/WebDriver/chromedriver_win32/chromedriver.exe'
driver = webdriver.Chrome(chrome_path)
driver.get(root)
time.sleep(4)

kill_iframe(driver)

text_on_search_bar(driver, 'next')

html = driver.page_source
soup = BeautifulSoup(html, 'lxml')
results = soup.findAll('li', class_='result-item')
print('총', len(results), '개의 결과를 찾았습니다.')
for idx, result in enumerate(results):
    print((idx+1), ':', result.get_text())    
sel = int(input('이 중 버노를 선택해주셈 -----> '))

clear_search(driver)
text_on_search_bar(driver, 'next')
result = driver.find_elements_by_class_name('result-item')
result[sel-1].click()
```
